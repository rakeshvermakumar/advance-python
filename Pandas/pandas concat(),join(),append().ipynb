# pandas concat method-use to combine more then one series and dataframes

```python
import pandas as pd
```

```python
se1=pd.Series([1,2,3,4,5])
se1
```

```python
se2=pd.Series([6,7,8,9,10])
se2
```

```python
pd.concat([se1,se2])
```

```python
se3=pd.Series([1,2,3,4,5])
se3
```

```python
se4=pd.Series([6,7,8])
se4
```

```python
pd.concat([se3,se4])
```

```python
df=pd.DataFrame({'id':[10,11,12,13,14],
              'Class':[1,2,3,4,5]})
df
```

```python
df1=pd.DataFrame({'id':[10,11,12,13,14],
              'name':['R','A','K','E','S']})
df1
```

```python
pd.concat([df,df1])
```

```python
pd.concat([df,df1],axis=1)
```

```python
pd.concat([df,df1],axis=0)
```

```python
pd.concat([df,df1],axis=0,ignore_index=True)
```

```python
pd.concat([df,df1],join='inner')
```

```python
pd.concat([df,df1],axis=1,join='inner')
```

```python
df1=pd.DataFrame({'id':[10,11,12,13,14],
              'Class':[1,2,3,4,5]}) 
df1
```

```python
df2=pd.DataFrame({'id':[15,16],
              'Class':[6,7],
                 'section':[8,9]})
df2
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join='inner')#only print common value(Excluded NaN value)
```

```python
pd.concat([df1,df2],axis=1,join='outer')#print all common values(included NaN values)
```

```python
pd.concat([df1,df2],axis=0,keys=['a','b'])
```

```python
pd.concat([df1,df2],axis=1,keys=['df','df1'])
```

# pandas join method

```python
df5=pd.DataFrame({'id':[10,11,12,13,14],
              'Class':[1,2,3,4,5]})

df6=pd.DataFrame({'roll':[15,16,17,18,19],
              'section':[6,7,8,9,10]})
display(df5,df6)
```

```python
df5.join(df6)
```

```python
df5=pd.DataFrame({'id':[10,11,12,13,14],
              'Class':[1,2,3,4,5]},index=['a','b','c','d','e'])

df6=pd.DataFrame({'roll':[15,16,17,18,19],
              'section':[6,7,8,9,10]})
display(df5,df6)
```

```python
df5.join(df6)
```

```python
df5=pd.DataFrame({'id':[10,11,12,13,14],
              'Class':[1,2,3,4,5]},index=['a','b','c','d','e'])

df6=pd.DataFrame({'roll':[15,16,17,18,19],
              'section':[6,7,8,9,10]},index=['a','b','c','d','e'])
display(df5,df6)
```

```python
df6.join(df5)
```

```python
df5=pd.DataFrame({'id':[10,11,12,13,14],
              'Class':[1,2,3,4,5]},index=['a','b','c','d','e'])

df6=pd.DataFrame({'roll':[15,16,19],
              'section':[6,7,10]},index=['a','b','c'])
display(df5,df6)
```

```python
df5.join(df6)
```

```python
df5.join(df6,how='right')
```

```python
df5.join(df6,how='left')
```

```python
df5.join(df6,how='inner')
```

```python
df5.join(df6,how='outer')
```

```python
df5=pd.DataFrame({'id':[10,11,12,13,14],
              'Class':[1,2,3,4,5]},index=['a','b','c','d','e'])

df6=pd.DataFrame({'id':[15,16,19],
              'section':[6,7,10]},index=['a','b','c'])
display(df5,df6)
```

```python
df6.join(df5)
```

```python
df6.join(df5,lsuffix='_4')
```

```python
df6.join(df5,rsuffix='_4')
```

# pandas append function

```python
df5=pd.DataFrame({'id':[10,11,12,13,14],
              'Class':[1,2,3,4,5]})

df6=pd.DataFrame({'roll':[15,16,17,18,19],
              'section':[6,7,8,9,10]})
display(df5,df6)
```

```python
df5.append(df6)
```

```python
df5.append(df6,ignore_index=True)
```
