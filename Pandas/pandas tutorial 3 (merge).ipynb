```python
#MERGE-use to merge multiple dataframe into single datafram by index and columns 
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'id':[10,11,12,13,14],
              'Class':[1,2,3,4,5]})
df
```

```python
df1=pd.DataFrame({'id':[10,11,12,13,14],
              'name':['R','A','K','E','S']})
df1
```

```python
a=pd.merge(df,df1,on='id')
a
```

```python
a=pd.merge(df,df1,on='id')
a
```

```python
b=pd.merge(df1,df,on='id')
```

```python
b
```

```python
df2=pd.DataFrame({'id':[10,11,12,13,14],
              'Class':[1,2,3,4,5]})
df2
```

```python
df4=pd.DataFrame({'id':[10,11,12,13,16],
              'name':['R','A','K','E','S']})
df4
```

```python
pd.merge(df2,df4)
```

```python
pd.merge(df2,df4,how='left')
```

```python
pd.merge(df2,df4,how='right')
```

```python
pd.merge(df2,df4,how='outer')
```

```python
pd.merge(df2,df4,how='outer',indicator=True)
```

```python
df3=pd.DataFrame({'id':[10,11,12,13,14],
              'Class':[1,2,3,4,5]})
df3
```

```python
df5=pd.DataFrame({'id':[15,16,17,18,19],
              'name':['R','A','K','E','S']})
df5
```

```python
pd.merge(df3,df5,left_index=True,right_index=True)
```

```python
df6=pd.DataFrame({'id':[15,16,17,18,19],
              'name':['R','A','K','E','S']})
df6
```

```python
df7=pd.DataFrame({'id':[15,16,17,18,19],
              'name':['R','A','K','E','S']})
df7
```

```python
pd.merge(df6,df7,on='id')
```

```python
pd.merge(df6,df7,on='id',suffixes=("_abcd","-rakesh"))
```

```python
pd.merge(df6,df7,sort=True)
```
