# pivot - The pivot() function is used to reshaped a given DataFrame organized by given index / column values.

```python
import pandas as pd
df=pd.read_csv('cars1.csv')
```

```python
df
```

```python
df.pivot_table(index="car")
```

```python
df.pivot_table(index="mpg")
```

```python
import pandas as pd
df=pd.read_csv('rakesh.csv')
```

```python
df
```

```python
df.pivot_table(index='avg',columns='years')
```

```python
df.pivot_table(index='avg',columns='years',aggfunc="count")
```

```python
df.pivot_table(index='avg',columns='years',aggfunc="max")
```

```python
df.pivot_table(index='avg',columns='years',aggfunc="sum",margins=True)
```

```python
df.pivot_table(index='avg',columns='years',fill_value=0)
```

```python
df.pivot_table(index='avg',columns='years',fill_value='none',margins=True)
```

```python
df.pivot_table(index='avg',columns='years',fill_value='none')
```

```python
df.pivot_table(index='avg',columns='years',aggfunc='sum',fill_value='none')
```

# Melt-pandas melt function is used to transform or reshape data

```python
M=pd.read_csv('rakesh.csv')
```

```python
M
```

```python
pd.melt(M)
```

```python
pd.melt(M,id_vars=['section'])
```

```python
pd.melt(M,id_vars=['years'])
```

```python
pd.melt(M,id_vars=['section'],value_vars=['marks'])
```

```python
pd.melt(M,id_vars=['section'],value_vars=['percentage'])
```

```python
pd.melt(M,id_vars=['section'],value_vars=['collage id'],var_name='collageid')
```

```python
pd.melt(M,id_vars=['section'],value_vars=['collage id'],var_name='school id',value_name='Data')
```
