```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1,color='b',linewidth=1,marker="o", markersize=10)#marker is use to design different line to see marker use (shift+tab)
plt.plot(x,y2,color='g',linestyle='-',linewidth=1, marker=">")
plt.title("line plot",fontsize = 15)
plt.xlabel("a",fontsize=15)
plt.ylabel("b",fontsize=29)
plt.legend(["abc"] ,loc=2)#use to write in side the graph
plt.grid(True)
```

```python
from matplotlib import style
x=np.arange(1,11)
y=2*x
style.use("ggplot")
plt.plot(x,y)
plt.title("line plot")
plt.plot(x,y1,color='m',linewidth=1,marker="o", markersize=10,label="abc")#marker is use to design different line to see marker use (shift+tab)
                                                                          #label is also use to write somthing in side the graph
plt.title("line plot",fontsize = 12)
plt.xlabel("a",fontsize=15)
plt.ylabel("b")
plt.legend(loc=4)#use to write in side the graph
plt.grid(color="g",linestyle="-.",linewidth=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
web_stud = np.random.randint(15,50,(50))
py_stud =np.random.randint(12,40,(50))
```

```python
print(web_stud)
print(py_stud)
```

```python
plt.hist(web_stud)
plt.title("webdevlopment student")
plt.xlabel("student age")
plt.ylabel("no.of student")
```

```python
plt.hist(
    x,
    bins=None,
    range=None,
    density=False,
    weights=None,
    cumulative=False,
    bottom=None,
    histtype='bar',
    align='mid',
    orientation='vertical',
    rwidth=None,
    log=False,
    color=None,
    label=None,
    stacked=False,
    *,
    data=None,
    **kwargs,
)
```

```python

plt.hist(web_stud,rwidth=0.8,histtype = "bar",orientation="horizontal",color="g",label="web_stud")
plt.title("webdevlopment student")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.legend()
```

```python
bins = [10,15,20,25,30,35,40,45]
plt.hist(web_stud,bins,rwidth=0.8,histtype = "bar",orientation="horizontal",color="g",label="web_stud")
plt.hist(py_stud,bins,rwidth=0.8,histtype = "bar",orientation="vertical",color="r",label="py_stud")
plt.title("webdevlopment student")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.legend()
```

```python
bins = [10,15,20,25,30,35,40,45]
plt.hist(web_stud,bins,rwidth=0.8,histtype = "bar",orientation="vertical",color="g",label="web_stud")
plt.hist(py_stud,bins,rwidth=0.8,histtype = "bar",orientation="vertical",color="r",label="py_stud")
plt.title("webdevlopment student")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.legend()
```

```python
bins = [15,20,25,30,35,40]
plt.hist([web_stud, py_stud],bins,rwidth=0.5,histtype = "bar",orientation="vertical",color=["g","b"],label=["web_stud","py_stud"])
#plt.hist(web_stud,bins,rwidth=0.8,histtype = "bar",orientation="vertical",color="r",label="py_stud")
plt.title("webdevlopment student")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.legend()
```

```python
bins = [15,20,25,30,35,40]
plt.figure(figsize = (16,8))
plt.hist([web_stud, py_stud],bins,rwidth=0.8,histtype = "bar",orientation="vertical",color=["g","b"],label=["web_stud","py_stud"])
#plt.hist(web_stud,bins,rwidth=0.8,histtype = "bar",orientation="vertical",color="r",label="py_stud")
plt.title("webdevlopment student")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.legend()
```

```python
from matplotlib import style
style.use("ggplot")
bins = [15,20,25,30,35,40]
plt.figure(figsize = (16,8))
plt.hist([web_stud, py_stud],bins,rwidth=0.8,histtype = "bar",orientation="vertical",color=["g","b"],label=["web_stud","py_stud"])
#plt.hist(web_stud,bins,rwidth=0.8,histtype = "bar",orientation="vertical",color="r",label="py_stud")
plt.title("webdevlopment student")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.legend()
plt.grid(color="r")
```
